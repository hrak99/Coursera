```python
total_days = 8
sellers = [(8,14), (15,11), (20,12), (25,13), (30,10)]
purchases=[]
purchases_price=[]

def calculate_purchasing_plan(total_days, sellers):
    loaves = 10   
    for days_seller1,price_seller1 in sellers[:1]:
        loaves=loaves-days_seller1
        buy_seller1 = days_seller1
        purchase_price=price_seller1 * days_seller1
        loaves=loaves+buy_seller1
        print ('Days from first seller arrives', days_seller1)
        print ("Buying Loaves from Seller1:",buy_seller1)
        print ("Total Purchase Price (in Pennies) from Seller1:",  purchase_price)
        purchases.append(loaves)
        purchases_price.append(purchase_price)
       
    for days_seller2,price_seller2 in sellers[1:2]:
        days_difference = days_seller2 - days_seller1
        loaves= loaves - days_difference
        buy_seller2 = 2*days_difference
        purchase_price=price_seller2 * (2*days_difference)
        loaves=loaves+buy_seller2
        print ('Days from second seller arrives', days_seller2)
        print ("Buying Loaves from Seller2:",buy_seller2)
        print ("Total Purchase Price (in Pennies) from Seller2:",  purchase_price)
        purchases.append(loaves)
        purchases_price.append(purchase_price)
    #print (purchases)
    
    for days_seller3,price_seller3 in sellers[2:3]:
        print ("Prices from periodically coming Sellers:",price_seller1, price_seller2, price_seller3)
        if price_seller3 < price_seller2 and price_seller1:
            days_difference = days_seller3 - days_seller2
            loaves= loaves - days_difference 
            buy_seller3 = days_difference
            purchase_price=price_seller3 * days_difference
            loaves=loaves+buy_seller3
            print ("Buying Loaves from Seller3:",loaves)
            print ("Total Purchase Price (in Pennies) from Seller3:",  purchase_price)
            purchases.append(loaves)
            purchases_price.append(purchase_price)
        else:
            print ("No Purchase due to high price")
            purchases.append(None)
            purchases_price.append(None)
    for days_seller4,price_seller4 in sellers[3:4]:
        print ("Prices from periodically coming Sellers:", price_seller1, price_seller2, price_seller3,price_seller4)
        if price_seller4 < price_seller3 and price_seller2 and price_seller1:
            days_difference = days_seller3 - days_seller2
            loaves= loaves - days_difference 
            buy_seller3 = days_difference
            purchase_price=price_seller3 * days_difference
            loaves=loaves+buy_seller3
            print ("Buying Loaves from Seller3:",loaves)
            print ("Total Purchase Price (in Pennies) from Seller3:",  purchase_price)
            purchases.append(loaves)
            purchases_price.append(purchase_price)
        else: 
            print ("No Purchase due to high price")
            purchases.append(None)
            purchases_price.append(None)
    

    for days_seller5,price_seller5 in sellers[4:5]:
        print ("Prices from periodically coming Sellers:", price_seller1, price_seller2, price_seller3,price_seller4,\
              price_seller5)
        if price_seller5 < price_seller4 or price_seller3 or price_seller2 or price_seller1 :
            days_difference = days_seller5 - days_seller2
            loaves= loaves - days_difference 
            buy_seller5 = 2*days_difference
            purchase_price=price_seller5 * (2*days_difference)
            print ("Buying Loaves from Seller3:",buy_seller5)
            loaves=loaves+buy_seller5            
            print ("Total Purchase Price (in Pennies) from Seller3:",  purchase_price)
            purchases.append(loaves)
            purchases_price.append(purchase_price)
        else: 
            print ("No Purchase due to high price")
            purchases.append(None)
            purchases_price.append(None)
    return ("Purchases from each seller", purchases,"Total Purchase Price from each Seller",purchases_price)
    

        
        
calculate_purchasing_plan(total_days, sellers)

```

## Linear Regression Model

```python
df.drop(['PRT_ID'], axis=1, inplace = True)
```

```python
df = pd.get_dummies(df)
```

```python
x = df.drop('SALES_PRICE', axis=1)
y= df['SALES_PRICE']
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
train_x, valid_x, train_y, valid_y = train_test_split(x, y, test_size = 0.3, random_state = 2)

train_x.shape, valid_x.shape, train_y.shape, valid_y.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_log_error
```

```python
lreg = LinearRegression()
lreg.fit(train_x, train_y)
```

### Model Evaluation - RMSLE

```python
pred_train = lreg.predict(train_x)
train_score = np.sqrt(mean_squared_log_error(train_y,pred_train))
```

```python
pred_test = lreg.predict(valid_x)
valid_score=np.sqrt(mean_squared_log_error(valid_y,pred_test))
```

```python
print('Training score:', train_score)
print('Validation score:', valid_score)
```
